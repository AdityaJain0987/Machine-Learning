```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y)
plt.xticks([1,2,3,4,5])         # x-axis arguments
plt.yticks([1,4,9,16,25])
plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.title("sine x")
plt.ylabel("result", )
plt.xlabel("argument")
plt.show()
```

```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.title("sine x", fontdict= {'fontname': 'Comic Sans MS', 'fontsize':20})
plt.ylabel("result", )
plt.xlabel("argument")
plt.show()
```

```python
x = [1,2,3]
y = [1,4,6]


plt.plot(x,y, label="2k", color = 'red', linewidth=8, marker =".", markersize=24, markeredgecolor='green')

plt.legend()

plt.show()
```

```python
y = [1,4,6]
x = [1,2,3]
#use shorthand notation
#fmt = '[color][marker][line]'


plt.plot(x,y,'r*--' ,label="2k")
plt.legend()
plt.show()
```

```python
y = [1,4,6]
x = [1,2,3]
#use shorthand notation
#fmt = '[color][marker][line]'
x2 = np.arange(0,6,1)
plt.plot(x2,x2**2, 'g.-', label="X^2")



plt.plot(x,y,'r*--' ,label="2k")
plt.legend()
plt.show()
```
